```python
import pandas as pd
# feature데이터

data = pd.read_csv('./data.csv')
data
```

```python
data.rename(columns={"interest rate": "interest_rate"}, inplace=True)
```

```python
cols = data.columns[data.columns.get_loc('unemployment'):]  # 'unemployment'부터 마지막 열까지의 컬럼들
data[cols] = data[cols].replace(',', '', regex=True).astype(float)
```

```python
import numpy as np
data['pred'] = np.expm1(data['pred'])
data['pred']
```

```python
data.to_sql(name='data', con=engine_obj, if_exists='append', index=False)
```
