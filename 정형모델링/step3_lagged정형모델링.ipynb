```python
lagged_df = pd.DataFrame()
lags = [1, 6, 12, 18]
# lags(1, 6, 12) 적용
for feature in df.columns[:-1]:
    for lag in lags:
      lagged_df[f"{feature}_lag_{lag}"] = df[feature].shift(lag)
```

```python
lagged_df = pd.concat([lagged_df, df[['interest_rate']]], axis=1)
lagged_df
```

```python
# 딕셔너리로 저장
data_dict = {}

for lag in lags:
    data_dict[f'data_{lag}'] = lagged_df[['interest_rate'] + [f"{feature}_lag_{lag}" for feature in df.columns[:-1]]].dropna()
```

```python
features = lagged_df.drop(columns=['interest_rate'])
labels = lagged_df['interest_rate']
```

```python
features
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

# 데이터 세트 분리

train_size = int(len(features) * 0.7)

X_train = features[:train_size]
X_test = features[train_size:]
y_train = labels[:train_size]
y_test = labels[train_size:]

X_train.shape
y_train.shape
X_test.shape
y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import lightgbm as lgb
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error

# 랜덤포레스트 모델
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# 예측
y_pred = rf_model.predict(X_test)

# RMSE
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse_rf:.4f}")

# MAE
mae_rf = mean_absolute_error(y_test, y_pred)
print(f"MAE: {mae_rf:.4f}")

# R²
train_score = rf_model.score(X_train, y_train)
test_score = rf_model.score(X_test, y_test)
print(f"훈련R²: {train_score}")
print(f"테스트R²: {test_score}")
```

```python
# XGBoost 모델
xgb_model = xgb.XGBRegressor()
xgb_model.fit(X_train, y_train)

# 예측
y_pred = xgb_model.predict(X_test)

# RMSE
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse_xgb:.4f}")

# MAE
mae_xgb = mean_absolute_error(y_test, y_pred)
print(f"MAE: {mae_xgb:.4f}")


# R²
train_score = xgb_model.score(X_train, y_train)
test_score = xgb_model.score(X_test, y_test)
print(f"훈련R²: {train_score}")
print(f"테스트R²: {test_score}")
```

```python
# LightGBM 모델
lgb_model = lgb.LGBMRegressor(random_state=42)
lgb_model.fit(X_train, y_train)

# 예측
y_pred = lgb_model.predict(X_test)

# RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse:.4f}")

# MAE
mae = mean_absolute_error(y_test, y_pred)
print(f"MAE: {mae:.4f}")

# R²
train_score = lgb_model.score(X_train, y_train)
test_score = lgb_model.score(X_test, y_test)
print(f"훈련R²: {train_score}")
print(f"테스트R²: {test_score}")
```

```python
# 랜덤포레스트 하이퍼 파라미터
rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

rf_grid_search = GridSearchCV(rf_model, rf_params, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
rf_grid_search.fit(X_train, y_train)

# 하이퍼 파라미터 튜닝
rf_best_model = rf_grid_search.best_estimator_

# 예측
y_pred = rf_best_model.predict(X_test)

# RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse:.4f}")

# MAE
mae = mean_absolute_error(y_test, y_pred)
print(f"MAE: {mae:.4f}")

# R²
train_score = rf_best_model.score(X_train, y_train)
test_score = rf_best_model.score(X_test, y_test)
print(f"훈련R²: {train_score}")
print(f"테스트R²: {test_score}")
```
